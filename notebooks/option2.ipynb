# Option 2.  Apply prediction on total profit

## Description

This approach based on hint fro the task decription. Accumulated values of LTVs are used to predict LTV60. Here LTVs will grow over time, which means extrapolation function will have higher slope then in option1

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

def get_ltvs_list(df):
    ltvs = []

    if df.Ltv1 > 0:
        ltvs.append(df.Ltv1)

    if df.Ltv2 > 0:
        ltvs.append(df.Ltv2)

    if df.Ltv3 > 0:
        ltvs.append(df.Ltv3)
    
    if df.Ltv4 > 0:
        ltvs.append(df.Ltv4)
    
    if df.Ltv5 > 0:
        ltvs.append(df.Ltv5)
    
    if df.Ltv6 > 0:
        ltvs.append(df.Ltv6)
    
    if df.Ltv7 > 0:
        ltvs.append(df.Ltv7)  
    
    return ltvs

def get_ltvs_list_users(df):
    ltvs = []

    if df.Ltv1 > 0:
        ltvs.append(df.Ltv1*df.Users)

    if df.Ltv2 > 0:
        ltvs.append(df.Ltv2*df.Users)

    if df.Ltv3 > 0:
        ltvs.append(df.Ltv3*df.Users)
    
    if df.Ltv4 > 0:
        ltvs.append(df.Ltv4*df.Users)
    
    if df.Ltv5 > 0:
        ltvs.append(df.Ltv5*df.Users)
    
    if df.Ltv6 > 0:
        ltvs.append(df.Ltv6*df.Users)
    
    if df.Ltv7 > 0:
        ltvs.append(df.Ltv7*df.Users)  
    
    return ltvs

def linearRegression(y):
    X = []
    for i in range(0, len(y)) :
        X.append([i])

    model = LinearRegression()
    model.fit(X, y)   

    yNew = model.predict([[59]])
    return yNew[0]
```

### CSV file

```python
data = pd.read_csv("../assets/data/test_data.csv")
data['non_zero_ltvs'] = data.apply(get_ltvs_list, axis=1)

data['Ltv60'] = data['non_zero_ltvs'].apply(linearRegression)
data
```

```python
groupCountry = data.groupby('Country')
groupCountry['Ltv60'].sum()/groupCountry.size()
```

### JSON File

```python
data = pd.read_json("../assets/data/test_data.json")
data['non_zero_ltvs'] = data.apply(get_ltvs_list_users, axis=1)

data['Ltv60'] = data['non_zero_ltvs'].apply(linearRegression)
data
```

```python
groupCountry = data.groupby('Country')
groupCountry['Ltv60'].sum()/groupCountry['Users'].sum()
```
